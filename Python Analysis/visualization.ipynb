```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../Dataset/crm_clean_master_data.csv")
```

## Number of deals close vs lost 


```python
sns.countplot(data=data, x = 'deal_outcome', color='blue')
plt.title('Deal Outcome Distribution')
plt.savefig("deals_distribution.png", dpi=300, bbox_inches="tight")
plt.show()
```

## Funnel Stage Drop-off


```python
stage_count = data['deal_stage'].value_counts()

stage_count.plot(kind='bar', color='blue')
plt.title('Deals by Funnel Stage')
plt.ylabel('Number of Deals')
plt.savefig("drop_off.png", dpi=300,bbox_inches="tight")
plt.show()
```

## Sales Cycle Analysis


```python
close_deals = data[data['deal_status'] == 'Closed']
sns.histplot(close_deals['sales_cycle_days'], color='blue', bins=30)
plt.title('Sales Cycle Distribution (Closed Deals)')
plt.savefig("Sales_Cycle_Analysis.png", dpi=300,bbox_inches="tight")
plt.show()
```

## Revenue by Sales Agent


```python
agent_revenue = (
    close_deals
    .groupby('sales_agent')['close_value']
    .sum()
    .sort_values(ascending=False)
)
agent_revenue.head(10).plot(kind='bar', color='blue')
plt.title('Top 10 Sales Agents by Revenue')
plt.savefig("revenue_by_sales_agent.png", dpi=300,bbox_inches="tight")
plt.show()
```
