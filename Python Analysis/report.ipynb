```python
import pandas as pd
```

```python
data = pd.read_csv("../Dataset/crm_clean_master_data.csv")
```

## Funnel Summary Report


```python
funnel_summary = (
    data.groupby('deal_stage')
    .agg(
        total_deals = ('opportunity_id', 'count'),
        total_revenue = ('close_value', 'sum'),
    )
    .reset_index()
)
funnel_summary.to_csv("funnel_summary,csv", index=False)
```

```python
funnel_summary
```

## Sales Agent Performance Table

```python
agent_performance = (
    data.groupby('sales_agent')
    .agg(
        deals_cloed = ('opportunity_id','count'),
        total_revenue = ('close_value', 'sum'),
        avg_cycle_days = ('sales_cycle_days', 'mean')
    )
    .reset_index()
)
agent_performance.to_csv("crm_sales_agent_perf.csv", index=False)
```
